## Task-3 : Car Price Prediction With Machine Learning

### Importing Required Modules 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
import sklearn.metrics as metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import seaborn as sns

import warnings 
warnings.filterwarnings("ignore")
```

### Data Collection & PreProccesing

```python
#loading the data from csv file to a pandas Dataframe
Data_frame=pd.read_csv("car_preddiction.csv")
Data_frame.head()
```

```python
#dispay the info of dataset
Data_frame.info()
```

```python
#checking the number of rows and columns in the dataframe
Data_frame.shape
```

### Generating Heat Map Using sns Package

```python
plt.figure(figsize=(12,8))
sns.heatmap(Data_frame.corr(numeric_only=True),annot=True)
plt.title("Correlation Between The Columns")
```

### Storing Independant Features in x Variable and Dependant Feature in y Variable

```python
x = Data_frame.drop(['price','CarName','fueltype','car_ID','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'], axis=1)
y = Data_frame['price'].values
print(f"x shape : {x.shape}  y shape : {y.shape}")
    
```

### Splitting The Data into Training Data & Test Data

```python
#for training 70%
#for test 30%
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.30,random_state=42)
print("X-train : ",x_train.shape)
print("X-test : ",x_test.shape)
print("y-train : ",y_train.shape)
print("y-test : ",y_test.shape)
```

### Building Linear Regression model

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#predicting result for x_test input
pred = model.predict(x_test)
```

### Checking Accuracy of model using mean square error, mean absolute error and r2 score

```python
print("MSE : ",(metrics.mean_squared_error(pred,y_test)))
print("MAE : ",(metrics.mean_absolute_error(pred,y_test)))
print("R2 Score : ",(metrics.r2_score(pred,y_test)))
    
```
